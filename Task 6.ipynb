# Decision Tree
Decision trees can be used for classification as well as regression problems. The name itself suggests that it uses a flowchart like a tree structure to show the predictions that result from a series of feature-based splits. It starts with a root node and ends with a decision made by leaves.

### Let's begin...


# Importing the Dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn import tree
from sklearn.model_selection import train_test_split
```

# The Iris Dataset
#### The Iris Dataset contains four features (length and width of sepals and petals) of 50 samples of three species of Iris (Iris setosa, Iris virginica and Iris versicolor). 

## Importing the Dataset

```python
dataset = pd.read_csv('H:/Datasets/Iris.csv')
dataset.shape
```

```python
dataset.head(5)
```

### The Target  Variable

```python
np.unique(dataset['Species'])
```

### Dropping the ID Column

```python
dataset = dataset.drop(['Id'] , axis=1)
```

# Visualizing the Dataset

```python
sns.set_style("whitegrid")
sns.FacetGrid(dataset, hue ="Species",height = 6).map(plt.scatter,'SepalLengthCm','SepalWidthCm').add_legend()
```

### Splitting the feature columns in the dataset [SepalLength, SepalWidth,PetalLength ,PetalWidth] and the target column [Species]

```python
X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:, -1].values
X[0:5,:]
X.shape
```

## Building a Decision Tree of the Dataset

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(X , Y)
```

## Visualizing the Decision Tree
<b>The Gini impurity</b> measure is one of the methods used in decision tree algorithms to decide the optimal split from a root node, and subsequent splits.
<b>Gini Impurity tells us what is the probability of misclassifying an observation.
Lower the Gini the better the split.</b>

“X[0]”, “X[1]”, “X[2]”,“X[3]” are the feature columns 

X[0] = SepalLengthCm	
X[1] = SepalWidthCm
X[2] = PetalLengthCm
X[3] = PetalWidthCm


```python
plt.figure(dpi=150)
tree.plot_tree(dtree)
plt.title('Decision Tree of Iris Dataset')
plt.show()
```

```python
fn=['SepalLengthCm' , 'SepalWidthCm' , 'PetalLengthCm' , 'PetalWidthCm']
cn=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(dtree,feature_names = fn, class_names=cn, filled = True);
```
